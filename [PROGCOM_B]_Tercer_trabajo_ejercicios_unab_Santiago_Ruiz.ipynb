```python
operaciones = ['+', '-', '*', '/']
continuar = 'si'

print("Calculadora Simple")

while continuar.lower() == 'si':
    print("\nOperaciones disponibles:")
    for i in range(len(operaciones)):
        print(f"{i+1}. {operaciones[i]}")

    num1 = float(input("\nIngrese el primer número: "))
    num2 = float(input("Ingrese el segundo número: "))
    opcion = int(input("Seleccione el número de operación (1-4): ")) - 1

    resultado = 0
    operacion_valida = True

    if opcion >= 0 and opcion < len(operaciones):
        if operaciones[opcion] == '+':
            resultado = num1 + num2
        elif operaciones[opcion] == '-':
            resultado = num1 - num2
        elif operaciones[opcion] == '*':
            resultado = num1 * num2
        elif operaciones[opcion] == '/':
            if num2 != 0:
                resultado = num1 / num2
            else:
                print("Error: No se puede dividir entre cero")
                operacion_valida = False
    else:
        print("Error: Operación no válida")
        operacion_valida = False

    if operacion_valida:
        print(f"\nResultado: {num1} {operaciones[opcion]} {num2} = {resultado}")

    continuar = input("\n¿Desea realizar otra operación? (si/no): ")

print("\n¡Gracias por usar la calculadora!")
```

```python
print("Calificaciones en Letras")

registros = []
num_estudiantes = int(input("¿Cuántos estudiantes desea evaluar?: "))

for i in range(num_estudiantes):
    print(f"\nEstudiante {i+1}")
    nombre = input("Nombre del estudiante: ")

    nota_valida = False
    while not nota_valida:
        nota = float(input("Ingrese la nota (0-100): "))
        if 0 <= nota <= 100:
            nota_valida = True
        else:
            print("Error: La nota debe estar entre 0 y 100")


    letra = ""
    if nota >= 90:
        letra = "A"
    elif nota >= 80:
        letra = "B"
    elif nota >= 70:
        letra = "C"
    elif nota >= 60:
        letra = "D"
    else:
        letra = "F"

    estudiante = {
        "nombre": nombre,
        "nota": nota,
        "letra": letra
    }
    registros.append(estudiante)


print("\nResultados de Calificaciones")
for i in range(len(registros)):
    estudiante = registros[i]
    print(f"\nEstudiante: {estudiante['nombre']}")
    print(f"Nota numérica: {estudiante['nota']}")
    print(f"Calificación en letra: {estudiante['letra']}")


total_estudiantes = len(registros)
calificaciones_count = {"A": 0, "B": 0, "C": 0, "D": 0, "F": 0}

for estudiante in registros:
    calificaciones_count[estudiante['letra']] += 1

print("\nEstadísticas por Calificación")
for letra in calificaciones_count:
    cantidad = calificaciones_count[letra]
    porcentaje = (cantidad / total_estudiantes) * 100
    print(f"Calificación {letra}: {cantidad} estudiantes ({porcentaje:.1f}%)")
```
