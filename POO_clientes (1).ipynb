```
import csv
from google.colab import files

def load_members_from_csv(filename):
    members = []
    try:
        with open(filename, mode='r', encoding='utf-8') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                row["Peso (kg)"] = float(row["Peso (kg)"])
                row["Altura (cm)"] = float(row["Altura (cm)"])
                row["Ritmo cardíaco"] = int(row["Ritmo cardíaco"])
                row["Días en gym"] = int(row["Días en gym"])
                row["Índice de masa corporal"] = float(row["Índice de masa corporal"])
                members.append(row)
    except FileNotFoundError:
        print(f"Archivo '{filename}' no encontrado. Por favor, asegúrese de tener este archivo en el entorno.")
        return None
    return members

def save_members_to_csv(filename, members):
    if not members:
        return
    headers = members[0].keys()
    with open(filename, mode='w', newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=headers)
        writer.writeheader()
        for member in members:
            writer.writerow(member)

def print_table(members, role):
    if role == 'administrador':
        headers = [
            "Cédula", "Nombre", "Teléfono", "Motivo", "Peso (kg)", "Altura (cm)",
            "Ritmo cardíaco", "Días en gym", "Membresía activa",
            "Índice de masa corporal", "Estado corporal", "Recomendación"
        ]
    elif role == 'entrenador':
        headers = [
            "Cédula", "Nombre", "Motivo", "Peso (kg)", "Altura (cm)",
            "Ritmo cardíaco", "Índice de masa corporal", "Estado corporal", "Recomendación"
        ]
    else:
        return

    col_widths = {h: len(h) for h in headers}
    for member in members:
        for h in headers:
            col_widths[h] = max(col_widths[h], len(str(member[h])))

    header_row = " | ".join(h.ljust(col_widths[h]) for h in headers)
    separator = "-+-".join("-" * col_widths[h] for h in headers)
    print(header_row)
    print(separator)

    for m in members:
        row = " | ".join(str(m[h]).ljust(col_widths[h]) for h in headers)
        print(row)

def find_member_by_cedula(members, cedula):
    for m in members:
        if m["Cédula"] == cedula:
            return m
    return None

def edit_member(member, role):
    print("\nEditando información del miembro:")
    if role == 'cliente':
        # Client can edit Nombre, Teléfono, Motivo
        for field in ["Nombre", "Teléfono", "Motivo"]:
            current = member[field]
            new_val = input(f"{field} (actual: {current}) - nuevo valor (ENTER para mantener): ").strip()
            if new_val:
                member[field] = new_val

    elif role == 'entrenador':
        # Trainer can edit Peso (kg) and Ritmo cardíaco
        for field in ["Peso (kg)", "Ritmo cardíaco"]:
            current = member[field]
            new_val = input(f"{field} (actual: {current}) - nuevo valor (ENTER para mantener): ").strip()
            if new_val:
                try:
                    if field == "Peso (kg)":
                        member[field] = float(new_val)
                    else:
                        member[field] = int(new_val)
                except ValueError:
                    print(f"Valor inválido para {field}. Se mantiene el valor original.")

    elif role == 'administrador':
        # Admin can edit all fields except Cédula
        editable_fields = [k for k in member.keys() if k != "Cédula"]
        for field in editable_fields:
            current = member[field]
            new_val = input(f"{field} (actual: {current}) - nuevo valor (ENTER para mantener): ").strip()
            if new_val:
                # Convert numeric fields appropriately
                if field in ["Peso (kg)", "Altura (cm)", "Índice de masa corporal"]:
                    try:
                        member[field] = float(new_val)
                    except ValueError:
                        print(f"Valor inválido para {field}, valor ignorado.")
                elif field in ["Ritmo cardíaco", "Días en gym"]:
                    try:
                        member[field] = int(new_val)
                    except ValueError:
                        print(f"Valor inválido para {field}, valor ignorado.")
                else:
                    member[field] = new_val

    else:
        print("Rol no reconocido para edición.")

def show_client_data_filtered(member):
    print(f"\nInformación disponible:")
    for k, v in member.items():
        if k not in {"Cédula", "Teléfono"}:
            print(f"{k}: {v}")

def main():
    filename = "gym_members_extended.csv"
    members = load_members_from_csv(filename)
    if members is None:
        return

    print("Bienvenido al sistema del gimnasio.")
    print("Ingrese su rol: administrador, entrenador, o cliente.")
    role = input("Rol: ").strip().lower()

    if role == 'administrador':
        print_table(members, role)
        cedula = input("Ingrese la cédula del miembro que desea editar: ").strip()
        member = find_member_by_cedula(members, cedula)
        if member:
            edit_member(member, role)
            save_members_to_csv(filename, members)
            print("Cambios guardados.")
        else:
            print("Cédula no encontrada.")
        files.download(filename)

    elif role == 'entrenador':
        print_table(members, role)
        cedula = input("Ingrese la cédula del miembro que desea editar: ").strip()
        member = find_member_by_cedula(members, cedula)
        if member:
            edit_member(member, role)
            save_members_to_csv(filename, members)
            print("Cambios guardados.")
        else:
            print("Cédula no encontrada.")
        files.download(filename)

    elif role == 'cliente':
        cedula = input("Ingrese su cédula: ").strip()
        member = find_member_by_cedula(members, cedula)
        if member:
            show_client_data_filtered(member)
            choice = input("¿Desea editar su Nombre, Teléfono o Motivo? (si/no): ").strip().lower()
            if choice == "si":
                edit_member(member, role)
                save_members_to_csv(filename, members)
                print("Cambios guardados.")
        else:
            print("Cédula no encontrada.")

    else:
        print("Rol no válido. Por favor reinicie e intente nuevamente.")

if __name__ == "__main__":
    main()

```
